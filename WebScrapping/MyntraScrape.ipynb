```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
# Function to extract Product Title
def get_Description(soup):

    try:
        Description = soup.find("span", attrs={"id":'productDescription'})
        Description_value = Description.text       
        Description_string = Description_value.strip()

    except AttributeError:
        Description_string = ""

    return Description_string

```

```python

if __name__ == '__main__':

    # add your user agent 
    HEADERS = ({'User-Agent':'', 'Accept-Language': 'en-US, en;q=0.5'})

    # The webpage URL
    URL = "https://www.Myntra.com/s?k=clothing+4&ref=nb_sb_noss_2"

    # HTTP Request
    webpage = requests.get(URL, headers=HEADERS)

    # Soup Object containing all data
    soup = BeautifulSoup(webpage.content, "html.parser")

    # Fetch links as List of Tag Objects
    links = soup.find_all("a", attrs={'class':'a-link-normal s-no-outline'})

    # Store the links
    links_list = []

    # Loop for extracting links from Tag Objects
    for link in links:
            links_list.append(link.get('href'))

    d = {"Description":[],'URL':[]}
    
    # Loop for extracting product details from each link 
    for link in links_list:
        new_webpage = requests.get("https://www.Myntra.com" + link, headers=HEADERS)

        new_soup = BeautifulSoup(new_webpage.content, "html.parser")

        # Function calls to display all necessary product information
        d['URL'].append(link)
        d['Description'].append(get_Description(new_soup))
        

    
    Myntra_df = pd.DataFrame.from_dict(d) 
    Myntra_df = Myntra_df.dropna()
    Myntra_df.to_csv("MyntraDataset.csv", header=True, index=False)
```
